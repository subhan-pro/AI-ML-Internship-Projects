Neural Network Basics
Artificial Neural Network (ANN) is a machine learning model inspired by the human brain. It consists of input, hidden, and output layers. Neural networks learn patterns from data and are used for classification and prediction tasks.

```python
import tensorflow as tf

print(tf.__version__)
```

```python
from tensorflow.keras.datasets import mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
```

```python
model = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(x_train, y_train, epochs=5)
```

```python
loss, accuracy = model.evaluate(x_test, y_test)

print("Test Accuracy:", accuracy)
```

```python
prediction = model.predict(x_test)

print(prediction[0])
```

```python
import numpy as np

print("Predicted Digit:", np.argmax(prediction[0]))
print("Actual Digit:", y_test[0])
```
