```python
print("Hello World")

name = "Subhan Tolat"
age = 18

print(name)
print(age)





# Integer
age = 18
print("Integer:", age, type(age))

# Float
height = 5.10
print("Float:", height, type(height))

# String
name = "Tirth"
print("String:", name, type(name))

# List
fruits = ["Apple", "Banana", "Mango"]
print("List:", fruits, type(fruits))

# Tuple
colors = ("Red", "Green", "Blue")
print("Tuple:", colors, type(colors))

# Set
numbers = {1, 2, 3, 4}
print("Set:", numbers, type(numbers))

# Dictionary
student = {
    "name": "meet",
    "age": 17,
    "course":"Computer Science"
}
print("Dictionary:", student, type(student))
```
