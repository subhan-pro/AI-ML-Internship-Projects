```python
from fastapi import FastAPI
import nest_asyncio
import uvicorn
from pyngrok import ngrok
```

```python
app = FastAPI()

@app.get("/")
def home():
    return {"message": "Welcome to ML Model API"}
```

```python
@app.get("/predict")
def predict(age: int):
    if age >= 18:
        result = "Adult"
    else:
        result = "Minor"

    return {
        "Age": age,
        "Prediction": result
    }
```
