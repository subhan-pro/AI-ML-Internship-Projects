Mean is the average value of a dataset.

Formula:
Mean = Sum of all values / Number of values

Example:
Data = [10, 20, 30, 40, 50]
Mean = (10+20+30+40+50)/5 = 30

```python
#practical example
data = [10, 20, 30, 40, 50]

mean = sum(data) / len(data)

print("Mean:", mean)
```

Median is the middle value of a sorted dataset.

Example:
[10,20,30,40,50]

Middle value = 30

```python
#practical example
import statistics

data = [10, 20, 30, 40, 50]

median = statistics.median(data)

print("Median:", median)
```

Variance measures how spread out the data is from the mean.

Low variance = values close together
High variance = values spread apart

```python
#practical example
import numpy as np

data = [10, 20, 30, 40, 50]

variance = np.var(data)

print("Variance:", variance)
```

Probability measures the chance of an event occurring.

Formula:

Probability = Favorable Outcomes / Total Outcomes

Example:
Rolling a dice

Probability of getting 3

= 1/6
= 0.1667

```python
favorable = 1
total = 6

probability = favorable / total

print("Probability:", probability)
```

Correlation shows the relationship between two variables.

+1 = Strong Positive Relationship
0 = No Relationship
-1 = Strong Negative Relationship

```python
import pandas as pd

data = {
    "Hours_Studied": [1, 2, 3, 4, 5],
    "Marks": [20, 40, 60, 80, 100]
}

df = pd.DataFrame(data)

print(df.corr())
```

A vector is a collection of numbers.

Example:
[1, 2, 3]

```python
import numpy as np

vector = np.array([1, 2, 3])

print(vector)
```

A matrix is a table of numbers arranged in rows and columns.

```python
matrix = np.array([
    [1, 2],
    [3, 4]
])

print(matrix)
```

```python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

print(A + B)
```
